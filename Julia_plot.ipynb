```julia
gr()
x = 0:0.0005:2

m = 3
c = 7

f = m .* x .+ c  # (f = 3 * x + 7)   # dot before the algebric expression is very important.

plot(x,f,label="Linear",legend=:topleft)
xaxis!("x")
yaxis!("f(x)")
title!("Equation of a Stright Line")
# savefig("Linear Plot.png") # for saving on drive
```

```julia
# Global Temperature vs Number of Pirates
global_temperature = [14.4,14.5,14.8,15.2,15.5,15.8]
numpirates = [45000,20000,15000,5000,400,17];
```

```julia
gr()
```

```julia
plot(numpirates,global_temperature,label="line")
scatter!(numpirates,global_temperature,label="point")
```

```julia
xaxis!("Number of Pirates")
yaxis!("Global Temperature (C) ")
title!("Influence of pirate population on Global Warming")
```
